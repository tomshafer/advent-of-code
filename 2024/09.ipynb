# Advent of Code 2024, Day 9

```python
from itertools import accumulate
from operator import add


def read_input(path):
    with open(path) as f:
        inp = [int(ch) for ch in f.readline().strip()]
        offsets = list(accumulate(inp, add, initial=0))
    data = [
        (i, start, start + size - 1, size)
        for i, (size, start) in enumerate(zip(inp[::2], offsets[::2]))
    ]
    spaces = [(data[i][2] + 1, d[1] - data[i][2] - 1) for i, d in enumerate(data[1:])]
    spaces = [s for s in spaces if s[1] > 0]
    return data, spaces


data, spaces = read_input("09-example.txt")
```
