# Advent of Code 2024, Day 7

```python
from operator import add, mul
```

```python
def read_input(path: str):
    eqns = {}
    with open(path) as f:
        for line in f:
            k, v = line.split(":")
            eqns[int(k)] = list(map(int, v.strip().split()))
    return eqns
```

### Part 1

```python
2**max(len(t) for t in read_input("07.txt").values())
```

```python
eqns = read_input("07.txt")
```

```python
nok, tok, oks = 0, 0, []
for total, inputs in eqns.items():
    n_ops = len(inputs) - 1
    for i in range(2**n_ops):
        ops = [{0: add, 1: mul}[(i // 2**o) % 2] for o in range(n_ops)]
        res = inputs[0]
        for i, op in enumerate(ops):
            res = op(res, inputs[i + 1])
        # print(total, inputs, res)
        if res == total:
            nok += 1
            tok += total
            oks += [total]
            break
nok, tok
```

### Part 2

```python
3 ** max(len(t) for t in read_input("07.txt").values())
```

```python
eqns = read_input("07.txt")
```

```python
def concat(a,b):
    return int(f"{a}{b}")
```

```python
nok, tok, oks = 0, 0, []
for total, inputs in eqns.items():
    n_ops = len(inputs) - 1
    for i in range(3**n_ops):
        ops = [{0: add, 1: mul, 2: concat}[(i // 3**o) % 3] for o in range(n_ops)]
        res = inputs[0]
        for i, op in enumerate(ops):
            res = op(res, inputs[i + 1])
        # print(total, inputs, res)
        if res == total:
            nok += 1
            tok += total
            oks += [total]
            break
nok, tok
```
