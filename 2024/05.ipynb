# Advent of Code 2025, Day 5

```python
def read_input(path):
    rules, pages = [], []
    with open(path) as f:
        process_rules = True
        for line in f:
            if line.strip() == "":
                process_rules = False
                continue
            if process_rules:
                rules += [list(map(int, line.split("|")))]
            else:
                pages += [list(map(int, line.strip().split(",")))]
    return rules, pages
```

```python
rules, pages = read_input("05.txt")
```

## Part 1

```python
rdict = {k: set(r for a, r in rules if a == k) for k in set(r for r, _ in rules)}
```

```python
ok = []
for p in pages:
    flag = True
    for i, x in enumerate(p):
        yvals = rdict.get(x, set()) & set(p)
        for y in yvals:
            if p.index(y) < i:
                flag = False
                break
    if flag:
        ok += [p]
```

```python
sum([x[(len(x) - 1) // 2] for x in ok])
```

## Part 2

```python
bad = [p for p in pages if p not in ok]
```

```python
def isok(p, rules):
    for l, r in rules:
        if l in p and r in p and p.index(l) > p.index(r):
            return False
    return True
```

```python
fixed = []
for p in bad:
    i = 0
    while True:
        lx, rx = rules[i % len(rules)]
        if lx in p and rx in p and (il := p.index(lx)) > (ir := p.index(rx)):
            p[il], p[ir] = rx, lx
        if isok(p, rules):
            fixed += [p]
            break
        i += 1
```

```python
sum([x[(len(x) - 1)//2] for x in fixed])
```
