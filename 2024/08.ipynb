# Advent of Code 2024, Day 8

```python
import itertools as it

import numpy as np
```

```python
def read_input(path):
    with open(path) as f:
        return np.array([[c for c in line.strip()] for line in f])
```

## Part 1

```python
grid = read_input("08.txt")
```

```python
ans = []
letters = np.unique(grid[grid != "."])

for ltr in letters:
    coords = np.stack(np.nonzero(grid == ltr)).T
    for a, b in it.combinations(coords, 2):
        v = b - a
        ans += [a - v, b + v]
```

```python
ans = np.stack(ans)
np.unique(ans[((ans >= 0) & (ans < grid.shape[0])).all(axis=1)], axis=0).shape
```

## Part 2

```python
ans = []
for ltr in letters:
    coords = np.stack(np.nonzero(grid == ltr)).T
    for a, b in it.combinations(coords, 2):
        v = b - a
        ans += [a, b]
        for m in range(1, max(grid.shape) + 1):
            ans += [a - m * v, b + m * v]
```

```python
ans = np.stack(ans)
np.unique(ans[((ans >= 0) & (ans < grid.shape[0])).all(axis=1)], axis=0).shape
```
