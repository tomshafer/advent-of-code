# Advent of Code 2024, Day 4

```python
import numpy as np
```

```python
with open("04.txt") as f:
    ws = np.array([list(line.strip()) for line in f])
```

```python
n = 0
xcoords = np.array(list(zip(*np.nonzero(ws == "X"))))
```

```python
def horiz(a, r, c):
    cmp = list("XMAS")
    n = 0
    if a.shape[1] >= c + 4 and (a[r, c : c + 4] == cmp).all():
        # print(" L-> R", r, c)
        n += 1
    if c >= 3 and (a[r, c - 3 : c + 1][::-1] == cmp).all():
        # print(" R-> L", r, c)
        n += 1
    return n


def vert(a, r, c):
    cmp = list("XMAS")
    n = 0
    if a.shape[0] >= r + 4 and (a[r : r + 4, c] == list("XMAS")).all():
        # print("T ->B ", r, c)
        n += 1
    if r >= 3 and (a[r - 3 : r + 1, c][::-1] == cmp).all():
        # print("B ->T", r, c)
        n += 1
    return n


def diag(a, r, c):
    cmp = list("XMAS")

    n = 0

    # TL -> BR
    if a.shape[0] >= r + 4 and a.shape[1] >= c + 4:
        z = np.array([a[r + o, c + o] for o in range(4)])
        if (z == cmp).all():
            # print("TL->BR", r, c)
            n += 1
    # TR -> BL
    if a.shape[0] >= r + 4 and c >= 3:
        z = np.array([a[r + o, c - o] for o in range(4)])
        if (z == cmp).all():
            # print("TR->BL", r, c)
            n += 1
    # BL -> TR
    if r >= 3 and a.shape[1] >= c + 4:
        z = np.array([a[r - o, c + o] for o in range(4)])
        if (z == cmp).all():
            # print("BL->TR", r, c)
            n += 1
    # BR -> TL
    if r >= 3 and c >= 3:
        z = np.array([a[r - o, c - o] for o in range(4)])
        if (z == cmp).all():
            # print("BR->TL", r, c)
            n += 1
    return n
```

```python
n = 0
for r, c in xcoords:
    h = horiz(ws, r, c)
    v = vert(ws, r, c)
    d = diag(ws, r, c)
    n += h + v + d

n
```

### Part 2

```python
n = 0
xcoords = np.array(list(zip(*np.nonzero(ws == "A"))))
```

```python
def diag2(a, r, c, dir=1):
    cmp = list("MAS")

    # TL to BR and vice versa
    if dir == 1:
        if a.shape[0] > r + 1 and a.shape[1] > c + 1 and r > 0 and c > 0:
            z = np.array([a[r + o, c + o] for o in range(-1, 2)])
            return (z == cmp).all() or (z[::-1] == cmp).all()

    # BL to TR and vice versa
    else:
        if a.shape[0] > r + 1 and a.shape[1] > c + 1 and r > 0 and c > 0:
            z = np.array([a[r - o, c + o] for o in range(-1, 2)])
            return (z == cmp).all() or (z[::-1] == cmp).all()

    return False
```

```python
n = 0
for r, c in xcoords:
    d1 = diag2(ws, r, c, 1)
    d2 = diag2(ws, r, c, -1)
    n += int(d1 and d2)

n
```
